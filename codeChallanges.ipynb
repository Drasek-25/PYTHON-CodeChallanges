## Longest Common Prefix
[Leetcode](https://leetcode.com/problems/longest-common-prefix/description/)

Easy

Write a function to find the longest common prefix string amongst an array of strings.
If there is no common prefix, return an empty string "".

**Example 1:**

Input: strs = ["flower","flow","flight"]
Output: "fl"
Example 2:

**Example 2:**

Input: strs = ["dog","racecar","car"]
Output: ""
Explanation: There is no common prefix among the input strings.

**Constraints:**

1 <= strs.length <= 200
0 <= strs[i].length <= 200
strs[i] consists of only lowercase English letters.

````python
def longest_common_prefix(strs):

    # Edge cases
    if len(strs) == 1: 
        return strs[0]
    if strs[0] == "": 
        return strs[0]

    #iterate indexes of first word.
    for i in range(0, len(strs[0])):
        # Set letter for comparrison from first word at eahc index
        letter = strs[0][i]
        
        # Iterates word in list after the first
        for j in range(1, len(strs)):
            # Catching words smaller than strs[0]
            if len(strs[j]) == i: 
                return strs[0][:i]
            # Catch when a non-match occurs and returning current position
            if letter != strs[j][i]:
                if i == 0: 
                    return ""
                else: 
                    return strs[0][:i]

    return strs[0][:i+1]    

import unittest

class TestLongestCommonPrefix(unittest.TestCase):
    def test_example1(self):
        self.assertEqual(longest_common_prefix(["flower","flow","flight"]), "fl")
    
    def test_example2(self):
        self.assertEqual(longest_common_prefix(["ab", "a"]), "a")
    
    def test_example3(self):
        self.assertEqual(longest_common_prefix(["flower","flower","flower","flower"]), "flower")

if __name__ == "__main__":
    unittest.main(argv=[''], exit=False)
````

## Length of Last Word

[Leetcode](https://leetcode.com/problems/length-of-last-word/description/)

Easy

Given a string s consisting of words and spaces, return the length of the last word in the string.

A word is a maximal substring consisting of non-space characters only.

**Example 1:**

Input: s = "Hello World"
Output: 5
Explanation: The last word is "World" with length 5.

**Example 2:**

Input: s = "   fly me   to   the moon  "
Output: 4
Explanation: The last word is "moon" with length 4.

**Example 3:**

Input: s = "luffy is still joyboy"
Output: 6
Explanation: The last word is "joyboy" with length 6.

**Constraints:**

1 <= s.length <= 104
s consists of only English letters and spaces ' '.
There will be at least one word in s.

````python
def lengthOfLastWord(s: str) -> int:  
    wordList = s.split()
    return len(wordList[-1])

import unittest

class TestLengthOfLastWord(unittest.TestCase):
    def test_example1(self):
        self.assertEqual(lengthOfLastWord("Hello World"), 5)
    
    def test_example2(self):
        self.assertEqual(lengthOfLastWord("   fly me   to   the moon  "), 4)
    
    def test_example3(self):
        self.assertEqual(lengthOfLastWord("luffy is still joyboy"), 6)

if __name__ == "__main__":
    unittest.main(argv=[''], exit=False)
````

## Add Two Numbers

[Leetcode](https://leetcode.com/problems/add-two-numbers/description/)

Medium

You are given two non-empty linked lists representing two non-negative integers. The digits are stored in reverse order, and each of their nodes contains a single digit. Add the two numbers and return the sum as a linked list.
You may assume the two numbers do not contain any leading zero, except the number 0 itself.

**Example 1:**

Input: l1 = [2,4,3], l2 = [5,6,4]

Output: [7,0,8]

Explanation: 342 + 465 = 807.


**Example 2:**

Input: l1 = [0], l2 = [0]

Output: [0]


**Example 3:**

Input: l1 = [9,9,9,9,9,9,9], l2 = [9,9,9,9]

Output: [8,9,9,9,0,0,0,1]

**Constraints:**

The number of nodes in each linked list is in the range [1, 100].

0 <= Node.val <= 9

It is guaranteed that the list represents a number that does not have leading zeros.

````python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

# Creating example linked lists
def createList(list):
    newLinkedList = ListNode(list[0])
    current = newLinkedList
    # Iterate through the stringed number adding the new numbers into the list for each index
    for num in list[1:]:
        current.next = ListNode(num)
        current = current.next

    return newLinkedList

def addTwoNumbers(l1, l2):
    #C reate strings to make the numbers out of linked list into a concatinated string
    totalNum1 = ""
    totalNum2 = ""

    # Specify head of first list, move through list concatinating string values of numbers in list
    current = l1
    while current is not None:
        totalNum1 += str(current.val)
        current = current.next

    #Do it again on second list
    current = l2
    while current is not None:
        totalNum2 += str(current.val)
        current = current.next

    # Take the both concatinated number strings, reverse both, convert to int, total them, 
    # convert to string to reverse the number, leave the number as a string to iterate over them
    totalNum3 = str(int(totalNum1[::-1]) + int(totalNum2[::-1]))[::-1]

    # Initialize a new linked list, then hold the current location on current
    # This is reduntant considering createList already exists, but I only have that function
    # so that this code works in both this notebook and Leetcode
    answer = ListNode(int(totalNum3[0]))
    current = answer
    #iterate through the stringed number adding the new numbers into the list for each index
    for num in totalNum3[1:]:
        current.next = ListNode(int(num))
        current = current.next

    #return answer as it is holding the head of the list, current is pointing at the last node
    return answer

# Since Leetcode needed a linkedlist as a return, we need a standalone function to make our answer readable here
def getAnswer(l3):
    answer = ""
    current = l3
    while current is not None:
        answer += str(current.val)
        current = current.next

    return answer

import unittest

class TestAddTwoNumbers(unittest.TestCase):
    def test_example1(self):
        l1 = createList([2, 4, 3])
        l2 = createList([5, 6, 4])
        self.assertEqual(getAnswer(addTwoNumbers(l1, l2)), "708")
    
    def test_example2(self):
        l1 = createList([0])
        l2 = createList([0])
        self.assertEqual(getAnswer(addTwoNumbers(l1, l2)), "0")
    
    def test_example3(self):
        l1 = createList([9, 9, 9, 9, 9, 9, 9])
        l2 = createList([9, 9, 9, 9])
        self.assertEqual(getAnswer(addTwoNumbers(l1, l2)), "89990001")

if __name__ == "__main__":
    unittest.main(argv=[''], exit=False)
````

## Who likes it?

[Codewars](https://www.codewars.com/kata/5266876b8f4bf2da9b000362/train/python)

You probably know the "like" system from Facebook and other pages. People can "like" blog posts, pictures or other items. We want to create the text that should be displayed next to such an item.

Implement the function which takes an array containing the names of people that like an item. It must return the display text as shown in the examples:


[]                                -->  "no one likes this"

["Peter"]                         -->  "Peter likes this"

["Jacob", "Alex"]                 -->  "Jacob and Alex like this"

["Max", "John", "Mark"]           -->  "Max, John and Mark like this"

["Alex", "Jacob", "Mark", "Max"]  -->  "Alex, Jacob and 2 others like this"


Note: For 4 or more names, the number in "and 2 others" simply increases.


````python
def likes(names):
    # There might be a way to write this with less code, but it would reduce
    # readability, and honestly what would be the point? the syntax is so specific.
    if len(names) == 0:
        return "no one likes this"
    elif len(names) == 1:
        return f"{names[0]} likes this"
    elif len(names) == 2:
        return f"{names[0]} and {names[1]} like this"
    elif len(names) == 3:
        # Willing to buy Oxford comma!
        return f"{names[0]}, {names[1]} and {names[2]} like this"
    else:
        return f"{names[0]}, {names[1]} and {len(names)-2} others like this"
    
import unittest

class TestLikes(unittest.TestCase):
    def test_example1(self):
        self.assertEqual(likes([]), "no one likes this")
    
    def test_example2(self):
        self.assertEqual(likes(["Peter"]), "Peter likes this")
    
    def test_example3(self):
        self.assertEqual(likes(["Jacob", "Alex"]), "Jacob and Alex like this")
    
    def test_example4(self):
        self.assertEqual(likes(["Max", "John", "Mark"]), "Max, John and Mark like this")
    
    def test_example5(self):
        self.assertEqual(likes(["Alex", "Jacob", "Mark", "Max"]), "Alex, Jacob and 2 others like this")

if __name__ == "__main__":
    unittest.main(argv=[''], exit=False)
````

## Printer Errors
[CodeWars](https://www.codewars.com/kata/56541980fa08ab47a0000040/train/python)

In a factory a printer prints labels for boxes. For one kind of boxes the printer has to use colors which, for the sake of simplicity, are named with letters from a to m.


The colors used by the printer are recorded in a control string. For example a "good" control string would be aaabbbbhaijjjm meaning that the printer used three times color a, four times color b, one time color h then one time color a...


Sometimes there are problems: lack of colors, technical malfunction and a "bad" control string is produced e.g. aaaxbbbbyyhwawiwjjjwwm with letters not from a to m.


You have to write a function printer_error which given a string will return the error rate of the printer as a string representing a rational whose numerator is the number of errors and the denominator the length of the control string. Don't reduce this fraction to a simpler expression.


The string has a length greater or equal to one and contains only letters from ato z.

````python
def printer_error(s):
    
    s = s.lower() 
    errors = 0
    
    for letter in s:
        if ord(letter) > 109:
            errors += 1
            
    return f"{errors}/{len(s)}"

import unittest

class TestPrinterError(unittest.TestCase):
    def test_example1(self):
        self.assertEqual(printer_error("aaaaaaaaaaaaaaaabbbbbbbbbbbbbbbbbbmmmmmmmmmmmmmmmmmmmxyz"), "3/56")
    
    def test_example2(self):
        self.assertEqual(printer_error("kkkwwwaaaaaaaaaaaaaabbbbbbbbbbbbbbbbbbmmmmmmmmmmmmmmmmmmmxyz"), "6/60")
    
    def test_example3(self):
        self.assertEqual(printer_error("kkkwwwaaaaaaaaaaaaaabbbbbbbbbbbbbbbbbbmmmmmmmmmmmmmmmmmmmxyzuuuuu"), "11/65")

if __name__ == "__main__":
    unittest.main(argv=[''], exit=False)
````

## Ones and Zeros

[CodeWars](https://www.codewars.com/kata/578553c3a1b8d5c40300037c/train/python)

Given an array of ones and zeroes, convert the equivalent binary value to an integer.


Eg: [0, 0, 0, 1] is treated as 0001 which is the binary representation of 1.

````python
def binary_array_to_number(arr):
    
    bin = ""
    
    for num in arr:
        bin += str(num)

    return int(bin, 2)


import unittest

class TestBinaryArrayToNumber(unittest.TestCase):
    def test_example1(self):
        self.assertEqual(binary_array_to_number([0, 0, 0, 1]), 1)
    
    def test_example2(self):
        self.assertEqual(binary_array_to_number([0, 0, 1, 0]), 2)
    
    def test_example3(self):
        self.assertEqual(binary_array_to_number([1, 1, 1, 1]), 15)
    
    def test_example4(self):
        self.assertEqual(binary_array_to_number([0, 1, 1, 0]), 6)

if __name__ == '__main__':
    unittest.main(argv=[''], verbosity=1, exit=False)
````

## Valid Braces

[CodeWars](https://www.codewars.com/kata/5277c8a221e209d3f6000b56/train/python)

Write a function that takes a string of braces, and determines if the order of the braces is valid. It should return true if the string is valid, and false if it's invalid.

This Kata is similar to the Valid Parentheses Kata, but introduces new characters: brackets [], and curly braces {}. Thanks to @arnedag for the idea!

All input strings will be nonempty, and will only consist of parentheses, brackets and curly braces: ()[]{}.

**What is considered Valid?**

A string of braces is considered valid if all braces are matched with the correct brace.



````python
def valid_braces(string):

    open_list = []
    bracket_dict ={
        ")" : "(",
        "}" : "{",
        "]" : "["
    }

    for bracket in string:
        
        if bracket == "(" or bracket == "{" or bracket == "[":
            open_list.append(bracket)
            
        else:
            if len(open_list) == 0:
                return False

            elif open_list[-1] == bracket_dict[bracket]:
                open_list.pop()
                
            else: 
                return False

    if len(open_list) == 0:
        return True
    
    else:
        return False

import unittest

class TestValidBraces(unittest.TestCase):
    def test_example1(self):
        self.assertTrue(valid_braces("(){}[]"))
    
    def test_example2(self):
        self.assertTrue(valid_braces("([{}])"))
    
    def test_example3(self):
        self.assertFalse(valid_braces("(}"))
    
    def test_example4(self):
        self.assertFalse(valid_braces("[(])"))
    
    def test_example5(self):
        self.assertFalse(valid_braces("[({})](]"))

if __name__ == "__main__":
    unittest.main(argv=[''], exit=False)
````

## Replace With Alphabet Position

[CodeWars](https://www.codewars.com/kata/546f922b54af40e1e90001da/train/python)

In this kata you are required to, given a string, replace every letter with its position in the alphabet.

If anything in the text isn't a letter, ignore it and don't return it.

"a" = 1, "b" = 2, etc.

````python
def alphabet_position(text):

    text = text.lower().replace(" ", "")
    answer = ""

    for letter in text:             
        num = ord(letter) - 96
        
        if 1 <= num <= 26:                 
            answer += str(num) + " "

    return answer[:-1]

import unittest

class TestAlphabetPosition(unittest.TestCase):
    def test_example1(self):
        self.assertEqual(alphabet_position("The sunset sets at twelve o' clock."), "20 8 5 19 21 14 19 5 20 19 5 20 19 1 20 20 23 5 12 22 5 15 3 12 15 3 11")
    
    def test_example2(self):
        self.assertEqual(alphabet_position("The narwhal bacons at midnight."), "20 8 5 14 1 18 23 8 1 12 2 1 3 15 14 19 1 20 13 9 4 14 9 7 8 20")

if __name__ == "__main__":
    unittest.main(argv=[''], exit=False)
````

## Tribonacci Sequence

[CodeWars](https://www.codewars.com/kata/556deca17c58da83c00002db/train/python)

Well met with Fibonacci bigger brother, AKA Tribonacci.

As the name may already reveal, it works basically like a Fibonacci, but summing the last 3 (instead of 2) numbers of the sequence to generate the next. And, worse part of it, regrettably I won't get to hear non-native Italian speakers trying to pronounce it :(

So, if we are to start our Tribonacci sequence with [1, 1, 1] as a starting input (AKA signature), we have this sequence:

```[1, 1 ,1, 3, 5, 9, 17, 31, ...]```

But what if we started with [0, 0, 1] as a signature? As starting with [0, 1] instead of [1, 1] basically shifts the common Fibonacci sequence by once place, you may be tempted to think that we would get the same sequence shifted by 2 places, but that is not the case and we would get:

```[0, 0, 1, 1, 2, 4, 7, 13, 24, ...]```

Well, you may have guessed it by now, but to be clear: you need to create a fibonacci function that given a signature array/list, returns the first n elements - signature included of the so seeded sequence.

Signature will always contain 3 numbers; n will always be a non-negative number; if n == 0, then return an empty array (except in C return NULL) and be ready for anything else which is not clearly specified ;)

````python
import unittest

def tribonacci(sig, n):

    # Edge Cases
    if n == 0: return []
    if n == 1: return [sig[0]]
    if n == 2: return sig[:2]
    if n == 3: return sig

    # Iterate towards n, appending the total of the last 3 nums in list
    for i in range(2, n-1):
        num = sig[i-2] + sig[i-1] + sig[i]
        sig.append(num)
        
    return sig

import unittest

class TestTribonacci(unittest.TestCase):
    def test_example1(self):
        self.assertEqual(tribonacci([1, 1, 1], 10), [1, 1, 1, 3, 5, 9, 17, 31, 57, 105])
    
    def test_example2(self):
        self.assertEqual(tribonacci([1, 2, 3], 10), [1, 2, 3, 6, 11, 20, 37, 68, 125, 230])
    
    def test_example3(self):
        self.assertEqual(tribonacci([300, 200, 100], 0), [])
    
    def test_example4(self):
        self.assertEqual(tribonacci([0.5, 0.5, 0.5], 30), [0.5, 0.5, 0.5, 1.5, 2.5, 4.5, 8.5, 15.5, 28.5, 52.5, 96.5, 177.5, 326.5, 600.5, 1104.5, 2031.5, 3736.5, 6872.5, 12640.5, 23249.5, 42762.5, 78652.5, 144664.5, 266079.5, 489396.5, 900140.5, 1655616.5, 3045153.5, 5600910.5, 10301680.5])
    
    def test_example5(self):
        self.assertEqual(tribonacci([3, 2, 1], 10), [3, 2, 1, 6, 9, 16, 31, 56, 103, 190])

if __name__ == "__main__":
    unittest.main(argv=[''], exit=False)
````

## Eureka Numbers
*Actual name is "Take a Number And Sum Its Digits Raised To The Consecutive Powers And ....¡Eureka!!*

[CodeWars](https://www.codewars.com/kata/5626b561280a42ecc50000d1/train/python)

The number 89 is the first integer with more than one digit that fulfills the property partially introduced in the title of this kata. What's the use of saying "Eureka"? Because this sum gives the same number: 
89 = 8<sup>1</sup> + 9<sup>2</sup>


The next number in having this property is 135:

See this property again:
135 = 1<sup>1</sup> + 3<sup>2</sup> + 5<sup>3</sup>

Examples
Let's see some cases (input -> output):

```1, 10  --> [1, 2, 3, 4, 5, 6, 7, 8, 9]```

```1, 100 --> [1, 2, 3, 4, 5, 6, 7, 8, 9, 89]```

If there are no numbers of this kind in the range [𝑎,𝑏] the function should output an empty list.

```90, 100 --> []```

````python
def sum_dig_pow(a, b):

    answer = []

    for cur_num in range(a, b + 1):

        # Catch all single digits ints 
        if cur_num <= 9:
            answer.append(cur_num)
            continue
        
        # Create a list with each number seperated
        exploded_num = []
        for single_num in str(cur_num):
            exploded_num.append(int(single_num))

        # Go through the list and exponentiate the number by its index + 1
        totaled_num = 0
        for i in range(len(exploded_num)):
            totaled_num += exploded_num[i] ** (i + 1)
        
        if cur_num == totaled_num:
            answer.append(cur_num)
        
    return answer


import unittest

class TestSumDigPow(unittest.TestCase):
    def test_example1(self):
        self.assertEqual(sum_dig_pow(1, 10), [1, 2, 3, 4, 5, 6, 7, 8, 9])
    def test_example2(self):
        self.assertEqual(sum_dig_pow(1, 100), [1, 2, 3, 4, 5, 6, 7, 8, 9, 89])
    def test_example3(self):
        self.assertEqual(sum_dig_pow(10, 89),  [89])
    def test_example4(self):
        self.assertEqual(sum_dig_pow(10, 100),  [89])
    def test_example5(self):
        self.assertEqual(sum_dig_pow(90, 100), [])
    def test_example6(self):
        self.assertEqual(sum_dig_pow(89, 135), [89, 135])
    def test_example7(self):
        self.assertEqual(sum_dig_pow(880, 2136), [1306, 1676])
    def test_example8(self):
        self.assertEqual(sum_dig_pow(285, 2890), [518, 598, 1306, 1676, 2427])

if __name__ == "__main__":
    unittest.main(argv=[''], exit=False)
````
